#Actividad 2

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

for i, ax in enumerate(axes.flat):
    group_data = df[df['group'] == i + 1]
    ax.scatter(group_data['x'], group_data['y'])
    ax.set_title(f'Group {i + 1}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')

plt.tight_layout()
plt.show()
```

#Actividad 3

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

fig, ax = plt.subplots(figsize=(10, 8))

for i in range(1, 5):
    group_data = df[df['group'] == i]
    X = group_data[['x']]
    y = group_data['y']
    model = LinearRegression().fit(X, y)
    y_pred = model.predict(X)

    ax.scatter(group_data['x'], group_data['y'], label=f'Group {i}')
    ax.plot(X, y_pred, label=f'Regression Line (Group {i})')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Regression Lines for All Groups')
ax.legend()
plt.show()
```
